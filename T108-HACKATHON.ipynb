```
import os
import platform
import subprocess
import socket
import psutil
from datetime import datetime

def print_header(title):
    print(f"\n{'=' * 20} {title} {'=' * 20}")

def run_command(command):
    try:
        output = subprocess.check_output(command, shell=True, text=True, stderr=subprocess.DEVNULL)
        return output.strip()
    except:
        return "Error or Not Available"

report = []

# Start scan
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
report.append(f"<h1>Windows Vulnerability and Network Scanner</h1><p><b>Generated:</b> {timestamp}</p>")

# System Info
print_header("System Information")
os_info = platform.platform()
print(f"Operating System: {os_info}")
report.append(f"<h2>System Information</h2><pre>{os_info}</pre>")

# .NET Versions
dotnet_output = run_command(r'reg query "HKLM\SOFTWARE\Microsoft\NET Framework Setup\NDP\v4\Full"')
print(dotnet_output)
report.append(f"<h3>.NET Versions</h3><pre>{dotnet_output}</pre>")

# Antivirus Check (Basic)
print_header("Antivirus Info")
av_path = r"C:\Program Files\Windows Defender\MsMpEng.exe"
av_status = "Windows Defender Installed" if os.path.exists(av_path) else "Antivirus not found"
print(av_status)
report.append(f"<h3>Antivirus Info</h3><pre>{av_status}</pre>")

# Windows Defender Settings
defender_settings = run_command('powershell "Get-MpPreference | Select DisableRealtimeMonitoring, ExclusionPath, ScanScheduleDay"')
print(defender_settings)
report.append(f"<h3>Windows Defender Settings</h3><pre>{defender_settings}</pre>")

# Installed Hotfixes
print_header("Installed Hotfixes")
hotfixes = run_command('powershell "Get-HotFix | Select HotFixID, Description, InstalledOn"')
print(hotfixes[:500])
report.append(f"<h3>Installed Hotfixes</h3><pre>{hotfixes}</pre>")

# Firewall Rules
print_header("Firewall Rules")
firewall_rules = run_command('netsh advfirewall firewall show rule name=all')
print(firewall_rules[:500])
report.append(f"<h3>Firewall Rules</h3><pre>{firewall_rules[:500]}...\n[Truncated]</pre>")

# Environment Variables
print_header("Environment Variables")
env_vars = "\n".join([f"{k}={v}" for k, v in os.environ.items()])
print(env_vars)
report.append(f"<h3>Environment Variables</h3><pre>{env_vars}</pre>")

# Hostname & IP
print_header("Network Info")
hostname = socket.gethostname()
ip_address = socket.gethostbyname(hostname)
print(f"Hostname: {hostname}")
print(f"Local IP: {ip_address}")
report.append(f"<h2>Network Information</h2><h3>Hostname</h3><pre>{hostname}</pre><h3>Local IP</h3><pre>{ip_address}</pre>")

# ARP Table
arp_table = run_command("arp -a")
print(arp_table)
report.append(f"<h3>ARP Table</h3><pre>{arp_table}</pre>")

# TCP Connections
print_header("Active TCP Connections")
tcp_conns = psutil.net_connections(kind='inet')
tcp_info = "\n".join([
    f"{c.laddr.ip}:{c.laddr.port} -> {c.raddr.ip}:{c.raddr.port}" if c.raddr else f"{c.laddr.ip}:{c.laddr.port} -> LISTENING"
    for c in tcp_conns[:10]
])
print(tcp_info)
report.append(f"<h3>Active TCP Connections</h3><pre>{tcp_info}</pre>")

# Autoruns
print_header("Autorun Entries")
autorun = run_command('reg query "HKCU\\Software\\Microsoft\\Windows\\CurrentVersion\\Run"')
print(autorun)
report.append(f"<h3>Autorun Entries</h3><pre>{autorun}</pre>")

# Running Services
print_header("Running Services")
services = run_command('sc query type= service state= all')
print(services[:500])
report.append(f"<h3>Running Services</h3><pre>{services[:500]}...\n[Truncated]</pre>")

# Save static HTML report
with open("windows_vuln_report.html", "w", encoding='utf-8') as f:
    f.write("<html><head><title>Windows Scanner</title></head><body>")
    f.writelines(report)
    f.write("</body></html>")

print_header("âœ… Scan Complete")
print("HTML report saved as 'windows_vuln_report.html'")
```
